# üìñ Informaci√≥n sobre el queratocono.

## ¬øQu√© es un queratocono? 


La c√≥rnea es la ventana transparente en forma de c√∫pula en la parte delantera del ojo. Enfoca la luz en el ojo. El queratocono es cuando la c√≥rnea se adelgaza y sobresale como un cono. El cambio de forma de la c√≥rnea hace que los rayos de luz se desenfoquen. Como resultado, la visi√≥n es borrosa y distorsionada, lo que dificulta las tareas diarias como leer y conducir.

> 

## ¬øQu√© causa el queratocono?

Los m√©dicos no saben con certeza por qu√© las personas pueden presentar queratocono. En algunos casos, parece ser gen√©tico (corre en la familia). Aproximadamente 1 de cada 10 personas con queratocono tiene un familiar que tambi√©n lo tiene. El queratocono es tambi√©n asociado con alergias oculares y frotaci√≥n excesiva de los ojos.

El queratocono suele aparecer en personas durante su adolescencia tard√≠a hasta principios de sus 20 a√±os. Los s√≠ntomas de la visi√≥n empeoran lentamente a lo largo de aproximadamente 10 a 20 a√±os.

# ü´ß Preprocesamiento de datos.

```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
from skimage.io import imread
from skimage.transform import resize
from skimage.color import rgb2gray
from skimage.filters import threshold_otsu
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

Carga de im√°genes y sus etiquetas correspondientes desde la direcci√≥n path de archivos de imagen especificados en un DataFrame.

```python
train_validation_path = pd.read_csv('train_validation_path.csv')
independent_test_path = pd.read_csv('independent_test_path.csv')
```

Entrenar y evaluar el modelo

```python
def remove_text_symbols(image):
    gray_image = rgb2gray(image)
    threshold = threshold_otsu(gray_image)
    binary_image = gray_image > threshold
    return binary_image
```

Carga y preprocesamiento de im√°genes

```python
def load_and_preprocess_image(path):
    image = imread(path)
    resized_image = resize(image, (224, 224))
    cleaned_image = remove_text_symbols(resized_image)
    return cleaned_image
```

Visualizaci√≥n de im√°genes preprocesadas

```python
def load_images(df):
    images = []
    labels = []
    for index, row in df.iterrows():
        patient_images = []
        for col in ['CT_A', 'EC_A', 'EC_P', 'Elv_A', 'Elv_P', 'Sag_A', 'Sag_P']:
            image_path = row[col]
            patient_images.append(load_and_preprocess_image(image_path))
        images.append(patient_images)
        labels.append(row['Label'])
    return images, labels
```

Normalizaci√≥n

```python
def normalize_images(images):
    normalized_images = []
    for image_set in images:
        normalized_set = [image / 255.0 for image in image_set]
        normalized_images.append(normalized_set)
    return normalized_images
```

Aplanado de im√°genes

```python
def flatten_images(images):
    flattened_images = []
    for image_set in images:
        flattened_set = [image.flatten() for image in image_set]
        flattened_images.append(np.concatenate(flattened_set))
    return np.array(flattened_images)
```

Escalado de datos

```python
def standardScaler(xTrain, xTest):
    scaler = StandardScaler()
    xTrain_scaled = scaler.fit_transform(xTrain)
    xTest_scaled = scaler.transform(xTest)
    return xTrain_scaled, xTest_scaled
```

Reducir dimensionalidad

```python
def reduce_dimension(images, n_components):
    pca = PCA(n_components=n_components)
    reduced_images = pca.fit_transform(images)
    return reduced_images, pca
```

```python
def train_and_evaluate_bayes(train_data, train_labels, test_data, test_labels):
    model = GaussianNB()
    model.fit(train_data, train_labels)
    train_predictions = model.predict(train_data)
    test_predictions = model.predict(test_data)
    train_accuracy = accuracy_score(train_labels, train_predictions)
    test_accuracy = accuracy_score(test_labels, test_predictions)
    train_report = classification_report(train_labels, train_predictions)
    test_report = classification_report(test_labels, test_predictions)
    train_conf_matrix = confusion_matrix(train_labels, train_predictions)
    test_conf_matrix = confusion_matrix(test_labels, test_predictions)
    return train_accuracy, test_accuracy, train_report, test_report, train_conf_matrix, test_conf_matrix
```

```python
def display_preprocessed_images(images, num_samples=5):
    fig, axes = plt.subplots(num_samples, 7, figsize=(15, 15))
    for i in range(num_samples):
        for j in range(7):
            axes[i, j].imshow(images[i][j], cmap='gray')
            axes[i, j].axis('off')
    plt.show()
```

```python
train_images, train_labels = load_images(train_validation_path)
test_images, test_labels = load_images(independent_test_path)
```

```python
train_images_normalized = normalize_images(train_images)
test_images_normalized = normalize_images(test_images)
```

# üïµÔ∏è Inspecci√≥n de los datos. 

```python
display_preprocessed_images(train_images_normalized)
```

Aplanamiento tanto para el conjunto de entrenamiento como para el de prueba.

```python
train_images_flattened = flatten_images(train_images_normalized)
test_images_flattened = flatten_images(test_images_normalized)
```

Aplicar StandardScaler

```python
xTrain_scaled, xTest_scaled = standardScaler(train_images_flattened, test_images_flattened)
```

# üóÉÔ∏è Clasificaci√≥n.

## M√©todo 1: Bayes Ingenuo. 

```python
n_components_range = range(1, 151, 15)


train_accuracies = []
test_accuracies = []
variances = []
train_conf_matrices = []
test_conf_matrices = []
```

Entrenamiento con reduccion de dimensiones.

```python
for n_components in n_components_range:

    reduced_train_images, pca = reduce_dimension(xTrain_scaled, n_components)
    
  
    reduced_test_images = pca.transform(xTest_scaled)
    

    train_accuracy, test_accuracy, train_report, test_report, train_conf_matrix, test_conf_matrix = train_and_evaluate_bayes(reduced_train_images, train_labels, reduced_test_images, test_labels)
    

    train_accuracies.append(train_accuracy)
    test_accuracies.append(test_accuracy)
    variances.append(np.sum(pca.explained_variance_ratio_))
    train_conf_matrices.append(train_conf_matrix)
    test_conf_matrices.append(test_conf_matrix)

    print(f"Components: {n_components}, Train Accuracy: {train_accuracy:.4f}, Test Accuracy: {test_accuracy:.4f}, Variance: {np.sum(pca.explained_variance_ratio_):.4f}")
    print("Train Confusion Matrix:\n", train_conf_matrix)
    print("Test Confusion Matrix:\n", test_conf_matrix)
    print("Train Classification Report:\n", train_report)
    print("Test Classification Report:\n", test_report)
```

Calcula la varianza explicada por cada componente

```python
plt.figure(figsize=(18, 12))

plt.subplot(2, 2, 1)
plt.plot(n_components_range, variances, marker='o')
plt.xlabel('Number of Components')
plt.ylabel('Explained Variance')
plt.title('Explained Variance vs Number of Components')

plt.subplot(2, 2, 2)
plt.plot(n_components_range, train_accuracies, label='Train Accuracy', marker='o')
plt.plot(n_components_range, test_accuracies, label='Test Accuracy', marker='o')
plt.xlabel('Number of Components')
plt.ylabel('Accuracy')
plt.title('Accuracy vs Number of Components')
plt.legend()

fig, axes = plt.subplots(len(n_components_range), 2, figsize=(20, 40))
fig.suptitle('Confusion Matrices for Different Number of PCA Components', fontsize=20)
for i, (train_conf_matrix, test_conf_matrix) in enumerate(zip(train_conf_matrices, test_conf_matrices)):
    sns.heatmap(train_conf_matrix, annot=True, fmt='d', cmap='Blues', ax=axes[i, 0])
    axes[i, 0].set_title(f'Train Confusion Matrix (Components: {n_components_range[i]})')
    axes[i, 0].set_xlabel('Predicted Label')
    axes[i, 0].set_ylabel('True Label')
    
    sns.heatmap(test_conf_matrix, annot=True, fmt='d', cmap='Blues', ax=axes[i, 1])
    axes[i, 1].set_title(f'Test Confusion Matrix (Components: {n_components_range[i]})')
    axes[i, 1].set_xlabel('Predicted Label')
    axes[i, 1].set_ylabel('True Label')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
